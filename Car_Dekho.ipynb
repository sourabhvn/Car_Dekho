# STEP 12
# HANDLING OUTLIERS



```python
import seaborn as sns
import matplotlib.pyplot as plt

#HANDLING OUTLIERS USING IQR METHOD
#Calculating Q1, Q3 and IQR for 'selling_price'

q1 = df['selling_price'].quantile(0.25)
q3 = df['selling_price'].quantile(0.75)
iqr = q3 - q1

#Defining lower and Upper limits
lower_cutoff = q1 - 1.5 * iqr
upper_cutoff = q3 + 1.5 * iqr

print("Lower Cutoff:", lower_cutoff)
print("Upper Cutoff:", upper_cutoff)

```

```python
#Checking Outliers
p0 = df['selling_price'].min()
p100 = df['selling_price'].max()
print(f"Min (P0): {p0}, Max (P100): {p100}")
```

```python
if lower_cutoff > p0:
    print("There are lower outliers")
else:
    print("No lower outliers")

if upper_cutoff < p100:
    print("There are upper outliers")
else:
    print("No upper outliers")
```

```python
#Checking using Boxplot
print("Boxplot BEFORE clipping:")
df.selling_price.plot(kind='box',figsize=(10,8),title='Selling Price (Before Clipping)')
```

WE CAN NOTICE THE OUTLIERS ON THE UPPER CUT OFF


NOW CLIPPING THE UPPER OUTLIERS USING WINSORIZATION

```python
df['selling_price'] = df['selling_price'].clip(upper=upper_cutoff)
```

```python
#Boxplotting after clipping
print("Boxplot AFTER clipping:")
df['selling_price'].plot(kind='box', figsize=(20, 14), title='Selling Price (After Clipping)')
```

WE CAN NOTICE THAT THE UPPER OUTLIERS ARE CLIPPED

# To enhance our analysis, we engineered a new column `value_tag` that classifies each car as either 'Valuable' or 'Not Valuable'.  
# A car is considered 'Valuable' if it is 5 years old or less and has a selling price above the dataset average.  
# This classification helps identify high-potential resale listings.  

```python
avg_price = df['selling_price'].mean()

df['value_tag'] = 'Not Valuable'  # default
df.loc[(df['car_age'] <= 5) & (df['selling_price'] > avg_price), 'value_tag'] = 'Valuable'
```

```python
valuable_cars_df = df[df['value_tag'] == 'Valuable']
print("Total Valuable Cars:", len(valuable_cars_df))
display(valuable_cars_df)
```
